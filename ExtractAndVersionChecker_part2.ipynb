```python
CARSEAT_DIR = "/media/dhruv2601/Carseat/Top500FreeAppDataset/"
app_ranking_file = os.path.join(CARSEAT_DIR, "app_ranking_500.txt")

app_ranking = open(app_ranking_file, "r")
apk_name = app_ranking.readlines()
```

```python
#get file names
DIR = "/media/dhruv2601/Carseat/Top500FreeAppDataset/data_extend/"
fileNames = apk_name

ind = 0

for i in range(0, len(fileNames)):
#     if (fileNames[i].split('.')[-1]) == 'apk':
        print(i)
        os.system('nodejs manifest_extractor_spider.js %s'%(os.path.join(DIR, fileNames[i])))

        file0 = json.load(open('ExtractorJSONoutput.json'))        

        if((file0!='{ }')):
            print(fileNames[i])

            file1 = file0['manifest']
            file2 = file1[0]
        
            packageName = file2['@package']
        
            file3 = file2['uses-sdk']
            file3 = file3[0]
            
            if('@android:targetSdkVersion' in file3):
                targetSDK = file3['@android:targetSdkVersion']
            else:
                targetSDK = -1
                
            if('@android:minSdkVersion' in file3):
                minSDK = file3['@android:minSdkVersion']
            else:
                minSDK = -1
                
            if('@android:maxSdkVersion' in file3):
                maxSDK = file3['@android:maxSdkVersion']
            else:
                maxSDK = -1
        
            file4 = file2['application'][0]
            
            if 'service' not in file4:
                noOfServices = 0
            else:
                file5 = file4['service']
                noOfServices = len(file5)
#                     for j in range(0, len(file5)):
#                         nameService.extend(file5[j]['@android:name'])
#                 print(nameService[j])
            df.loc[ind+i] = [i, packageName, minSDK, targetSDK, maxSDK, noOfServices]

df.shape[0]
```
